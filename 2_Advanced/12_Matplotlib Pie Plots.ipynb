```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## ðŸŸ© Health Insurance Pie Chart (2.14.1)

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()

insurance = df_DA['job_health_insurance'].value_counts()
plt.pie(insurance, labels=['No','Yes'], autopct = '%1.1f%%', startangle=90)
plt.title('Proportion of Data Analyst Jobs Mentioning Health Insurance')
```

---

## ðŸŸ¨ Job Schedule Distribution (2.14.2)

```python
df_ft_pt = df[df['job_schedule_type'].isin(['Full-time', 'Part-time'])]

schedule_type_counts = df_ft_pt['job_schedule_type'].value_counts()

plt.pie(schedule_type_counts, labels=['Full-time','Part-time'], autopct='%1.1f%%', startangle=90 )
plt.title('Job Schedule Full-time vs. Part-time')
```

---

## ðŸŸ¥ Degree Mentioned Pie Chart (2.14.3)

```python
df_US = df[df['job_country'] == 'United States' ].copy()

us_jobs_degree_mention = df_US['job_no_degree_mention'].value_counts()

plt.pie(us_jobs_degree_mention,labels=['Degree Mentioned','No Degree Mentioned'], autopct='%1.1f%%',startangle=90,explode=(0,0.1))
plt.title('Distribution of Jobs Mentioning Degrees for US Job Postings')
```
