```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Drop Missing Values (2.3.1)

```python
df_new = df.copy()

df_cleaned = df_new.dropna(subset = ['salary_year_avg'])

df_cleaned.head(5)
```

---

## ðŸŸ© Random Sample of DataFrame (2.3.2)

```python
df_sampled = df.sample(frac=0.1)
df_sampled.head(5)
```

## ðŸŸ¨ Fill Missing Salaries (2.3.3)

```python
df_filled = df.copy()

median_salary = df_filled['salary_year_avg'].median()

df_filled = df_filled['salary_year_avg'].fillna(median_salary)
df_filled
```
