```python
import pandas as pd

df = pd.read_csv("hf://datasets/lukebarousse/data_jobs/data_jobs.csv")
```

## ðŸŸ© Bar Chart with Labels (1.30.1)

```python
import matplotlib.pyplot as plt
```

```python
jobs_per_country = df.job_country.value_counts().head(5)

jobs_per_country
```

```python
plt.bar(jobs_per_country.index, jobs_per_country.values,color = 'orange')
plt.xlabel('Country')
plt.ylabel('Number Of Postings')
plt.title('Top 5 Countries With The Most Postings')
plt.xticks(rotation=45, ha='right')
```

---

## ðŸŸ¨ Horizontal Bar Chart with Labels (1.30.2)

```python
avg_yearly_sal = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=False)
avg_yearly_sal
```

```python
avg_yearly_sal = avg_yearly_sal.sort_values()
plt.barh(avg_yearly_sal.index,avg_yearly_sal)
plt.xlabel('Average Yearly Salary')
plt.title('Average Yearly Salary By Job Title')
```

## ðŸŸ¥ Line Chart with Labels (1.30.3)

```python
job_posted_date = df.job_posted_date
job_posted_date
```

```python
df.job_posted_date = pd.to_datetime(df.job_posted_date)

df['job_posted_month'] = df.job_posted_date.dt.month
monthly_avg_sal = df.groupby('job_posted_month')['salary_year_avg'].mean().sort_index()
monthly_avg_sal
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(monthly_avg_sal.index,monthly_avg_sal.values,color='red', label = 'Average Salary')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary By Month')
plt.legend()
```
