```python
import pandas as pd

df = pd.read_csv("hf://datasets/lukebarousse/data_jobs/data_jobs.csv")
```

## ðŸŸ© Job Country Bar Chart (1.29.1)

```python
import matplotlib.pyplot as plt
```

```python
jobs_per_country = df.job_country.value_counts().head(5)

jobs_per_country
```

```python
plt.bar(jobs_per_country.index, jobs_per_country.values,color = 'orange')
```

---

## ðŸŸ¨ Average Salary Horizontal Bar Chart (1.29.2)

```python
avg_salary = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
avg_salary
```

```python
plt.barh(avg_salary.index, avg_salary.values)
```

## ðŸŸ¥ Average Salary Trend Line Plot (1.29.3)

```python
df.job_posted_date = pd.to_datetime(df.job_posted_date)
df.info()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
df
```

---

```python
salary_per_month = df.groupby('job_posted_month')['salary_year_avg'].mean().sort_index()
salary_per_month
```

```python
plt.plot(salary_per_month.index, salary_per_month.values, color = 'red')
```
