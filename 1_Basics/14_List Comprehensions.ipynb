```python
# Code Needed
salaries_usd = [100000, 120000, 80000, 90000]

Convert_salaries = [salary*0.85 for salary in salaries_usd]
print(Convert_salaries)
```

## ðŸŸ¨ Create List of Data Jobs (1.14.2)

```python
# Code Needed
job_experience = {'Data Scientist': 3, 'Data Analyst': 1, 'Machine Learning Specialist': 4, 'Data Engineer': 2}

data_jobs = [job for job, experience in job_experience.items() if experience > 2]
print(data_jobs)
```

## ðŸŸ¥ Extract High Salary Jobs (1.14.3)

```python
# Code Needed
job_salary_dicts = [
    {'job_title': 'Data Scientist', 'salary': 100000},
    {'job_title': 'Data Analyst', 'salary': 120000},
    {'job_title': 'Data Engineer', 'salary': 80000},
    {'job_title': 'Machine Learning Engineer', 'salary': 90000}]

high_salary_jobs = [job['job_title'] for job in job_salary_dicts if job['salary'] > 90_000 ]
print(high_salary_jobs)


```
