```python
import pandas as pd

df = pd.read_csv('/content/data_jobs.csv', on_bad_lines='skip')  # λειτουργεί με pandas >= 1.3.0
  # παραλείπει τις "σπασμένες" γραμμές

```

```python
df
```

```python
df.info()
```

---

## 🟩 Sort by Salary (1.25.1)

```python
df.sort_values(by = 'salary_year_avg',ascending=False,inplace=True)
df.head(10)
```

---

## 🟩 Create Remote Column (1.25.2)

```python
df['Remote_work'] = df.job_work_from_home
df
```

---

## 🟩 Extract Year from Date (1.25.3)

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df.info()
```

```python
df['job_posted_year'] = df.job_posted_date.dt.year
df
```

---

## 🟨 Drop Missing Salaries (1.25.4)

```python
df.salary_year_avg
```

```python
df.dropna(subset= ['salary_year_avg'],inplace=True)
df
```
